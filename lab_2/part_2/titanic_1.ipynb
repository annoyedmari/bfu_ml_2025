```python
import pandas
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

file = "../Titanic.csv"
dataframe = pandas.read_csv(file)
```

```python
# Task 1
dataframe_clean = dataframe.dropna() # 1.1 deleting empty rows

columns_to_delete = ['Name', 'Ticket', 'Cabin'] # 1.2
dataframe_clean = dataframe_clean.drop(columns = columns_to_delete)

dataframe_clean['Sex'] = dataframe_clean['Sex'].map({'male': 0, 'female': 1}) # 1.3
dataframe_clean['Embarked'] = dataframe_clean['Embarked'].map({'C': 1, 'Q': 2, 'S': 3})

rows = len(dataframe)
rows_clean = len(dataframe_clean)
data_loss = ((rows - rows_clean) / rows) * 100 # 1.5 data loss after completing 1.1 and 1.2
# !! 1.4 not mentioned in 1.5 so PassengerId is removed after calculating data_loss
print(data_loss, '% of data is lost after completing 1.1 and 1.2.')
```

```python
dataframe_clean = dataframe_clean.drop(columns = 'PassengerId') # 1.4
```

```python
# Task 2 (Machine Learning)

def train_predict(X,Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42) # 2.1
    
    clf = LogisticRegression(max_iter=1000) # 2.2
    clf.fit(X_train, Y_train)
    
    Y_predict = clf.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_predict)
    
    return accuracy

X_0 = dataframe_clean.drop('Survived', axis = 1)
Y_0 = dataframe_clean['Survived']

accuracy_0 = train_predict(X_0, Y_0)
print("Accuracy score: ", accuracy_0) # 2.3

# 2.4 'Embarked' influence: data without embarked marked as X_1
X_1 = X_0.drop('Embarked', axis = 1)
accuracy_1 = train_predict(X_1, Y_0)
print("Accuracy score (without Embarked): ", accuracy_1) # 2.3

accuracy_diff = accuracy_0 - accuracy_1
print(f'Accuracy score diffence: {accuracy_0} - {accuracy_1} = {accuracy_diff}')
```
